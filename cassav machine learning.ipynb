```python
plt.figure(figsize=(10,10))
for i,col in enumerate(cont_data):
    plt.subplot(10,10,i+1)
    plt.boxplot(fem[col])
    plt.title(col)
    plt.ylabel(f"frequency")
plt.tight_layout()
plt.show()    
```

```python
def remove_outliers(columns):
    for column in columns:
        q1=fem[column].quantile(.25)
        q3=fem[column].quantile(.75)
        IQR=q3-q1
        upper_bound=q3+IQR*1.5
        lower_bound=q1-IQR*1.5
        fem[column]=np.where(fem[column]>upper_bound,upper_bound,fem[column])
        fem[column]=np.where(fem[column]<lower_bound,lower_bound,fem[column])
```
